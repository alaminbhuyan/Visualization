```python
import requests
from bs4 import BeautifulSoup
```

```python
# Our target url

url = "https://www.nbcnews.com/news/world/sanctions-force-russia-produce-popular-car-safety-features-even-kremli-rcna32863"
```

```python
response = requests.get(url=url)
```

```python
# Check the response

response
```

```python
# initialization the parser

soup = BeautifulSoup(markup=response.text, features='html.parser')

# soup = BeautifulSoup(markup=response.content, features='html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
target_content = soup.find(name='div', attrs={'class' : 'article-body__content'})
```

```python
target_content
```

```python
myArticle = ''

for i in target_content.find_all(name='div'):
    myArticle = myArticle + ' ' + i.text

    
print(myArticle)
```
