### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
```

```python
print("matplotlib version: ", matplotlib.__version__)
print("seaborn version: ", sns.__version__)
print("plotly version: ", plotly.__version__)

# matplotlib version:  3.6.3
# seaborn version:  0.12.2
# plotly version:  5.11.0
```

```python
tips = sns.load_dataset('tips')

tips
```

```python
sns.set_theme()
```

doc: https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot

```python
# scatter plot -> axes level function

# here, we visulize 5 columns informations

plt.figure(figsize=(7,6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
plt.show()
```

Doc: https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
# relplot -> figure level -> square shape

sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
plt.show()
```

```python
# line plot

gap = px.data.gapminder()

gap.head()
```

```python
temp_df = gap[gap['country'] == 'India']

temp_df
```

Doc: https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot

```python
# axes level function

sns.lineplot(data=temp_df, x='year', y='lifeExp')
plt.show()
```

Doc: https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
# line plot using relpplot

sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
plt.show()
```

```python
temp_df = gap[gap['country'].isin(values=['India','Brazil','Germany'])]

temp_df
```

```python
# hue -> style

sns.relplot(kind='line', data=temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
plt.show()
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
plt.show()
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot

sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
plt.show()
```

```python
# col wrap

sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=3)
plt.show()
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal? [means number of pick points]
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot

# axes level -> histplot -> kdeplot -> rugplot
```

Doc: https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

```python
# plotting univariate histogram

sns.histplot(data=tips, x='total_bill')
plt.show()
```

Doc: https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot

```python
# Using figure level function

sns.displot(data=tips, x='total_bill', kind='hist')
plt.show()
```

```python
# bins parameter

sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
plt.show()
```

```python
# bins parameter

sns.displot(data=tips, x='total_bill', kind='hist', bins=3)
plt.show()
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# workl like countplot

sns.displot(data=tips, x='day', kind='hist') 
plt.show()
```

```python
# hue parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
plt.show()
```

```python
# element -> step

sns.displot(data=tips, x='tip', kind='hist',hue='sex', element='step')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', hue='sex', kind='hist', element='step')
plt.show()
```

```python
# faceting using col and row -> not work on histplot function

sns.displot(data=tips, x='tip', kind='hist', col='sex', element='step')
plt.show()
```

Doc: https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, 
# producing a continuous density estimate

sns.kdeplot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', kind='kde')
plt.show()
```

```python
# hue -> fill

sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, height=10, aspect=2)
plt.show()
```

Doc: https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing 
# the location of individual observations in an unobtrusive way.

sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
plt.show()
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
plt.show()
```

Doc: https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

sns.kdeplot(data=tips, x='total_bill', y='tip')
plt.show()
```

### 2. Matrix Plot

- Heatmap
- Clustermap

Doc: https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn.heatmap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country', columns='year', values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
plt.show()
```

```python
# annot

temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df, annot=True, linewidth=0.5, cmap='summer')
plt.show()
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()

iris
```

Doc: https://seaborn.pydata.org/generated/seaborn.clustermap.html#seaborn.clustermap

```python
sns.clustermap(data=iris.iloc[:, [0,1,2,3]])
plt.show()
```
