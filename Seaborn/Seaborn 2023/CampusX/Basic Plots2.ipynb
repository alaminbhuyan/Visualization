### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

doc: https://seaborn.pydata.org/generated/seaborn.boxplot.html#seaborn.boxplot

```python
# Box plot

sns.boxplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
# Box plot

sns.boxplot(data=tips, x='day', y='total_bill', hue="sex")
plt.show()
```

```python
# Box plot

sns.boxplot(data=tips, x='total_bill', y='day', orient='h')
plt.show()
```

```python
# Using catplot

# Figure level function

sns.catplot(data=tips, x='day', y='total_bill', kind='box')
plt.show()
```

```python
# hue

sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
# single boxplot -> numerical col

sns.boxplot(data=tips, y='total_bill')
plt.show()
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

doc: https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot

```python
# axes level

# hue parameter is not available

sns.regplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

doc: https://seaborn.pydata.org/generated/seaborn.residplot.html#seaborn.residplot

```python
# residplot

sns.residplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# lmplot
# Figure level function

sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
# lmplot
# Figure level function

sns.lmplot(data=tips, x="total_bill", y="tip", hue="sex", col="day", row="time")
plt.show()
```

### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot

sns.catplot(data=tips, x='sex', y='total_bill', kind='violin', col='day', row='time')
plt.show()
```

doc: https://seaborn.pydata.org/generated/seaborn.FacetGrid.html#seaborn.FacetGrid

```python
g = sns.FacetGrid(data=tips, col="time",  row="sex")
g.map_dataframe(sns.histplot, x="total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map_dataframe(sns.histplot, x="total_bill", binwidth=2, binrange=(0, 60))
```

```python
g = sns.FacetGrid(tips, col="time", hue="sex")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

doc: https://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot

```python
sns.pairplot(data=iris)
plt.show()
```

```python
sns.pairplot(data=iris, hue="species")
plt.show()
```

doc: https://seaborn.pydata.org/generated/seaborn.PairGrid.html#seaborn.PairGrid

```python
# pair grid

g = sns.PairGrid(data=iris, hue='species')

g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot

doc: https://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot

```python
# jointplot
# ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'] you can pass this kinds

sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
# jointplot


sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

doc: https://seaborn.pydata.org/generated/seaborn.JointGrid.html#seaborn.JointGrid

```python
penguins = sns.load_dataset("penguins")
sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip')

g.plot(sns.kdeplot, sns.violinplot)
```

```python
g = sns.JointGrid(data=tips, x='total_bill', y='tip', hue="sex")

g.plot(sns.scatterplot, sns.histplot)
plt.show()
```
