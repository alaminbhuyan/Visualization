```python
# if we want to see scatter plot for respect to specific column than use hue. it return density curves 
sns.jointplot(x="bill_length_mm", y ="bill_depth_mm",data=penguins,hue="species")
plt.show()
```

#### Several different approaches to plotting are available through the kind parameter.
#### Setting kind="kde" will draw both bivariate and univariate KDEs

```python
# by default kind have "scatter"
sns.jointplot(x="bill_length_mm", y ="bill_depth_mm",data=penguins,hue="species",kind="kde")
plt.show()
```

#### Set kind="reg" to add a linear regression fit (using regplot()) and univariate KDE curves:

```python
sns.jointplot(x="bill_length_mm", y ="bill_depth_mm",data=penguins,hue="species",kind="hist")
plt.show()
```

#### Alternatively, setting kind="hex" will use matplotlib.axes.Axes.hexbin() to compute a bivariate histogram using hexagonal bins

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="hex")
plt.show()
```

#### Set kind="reg" to add a linear regression fit (using regplot()) and univariate KDE curves:

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", kind="reg")
plt.show()
```

```python
sns.jointplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    marker="+", s=100, marginal_kws=dict(bins=25, fill=False),
)
plt.show()
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", height=5, ratio=2, marginal_ticks=True)
plt.show()
```

```python
g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.kdeplot, color="r", zorder=0, levels=6)
g.plot_marginals(sns.rugplot, color="r", height=-.15, clip_on=False)
plt.show()
```
