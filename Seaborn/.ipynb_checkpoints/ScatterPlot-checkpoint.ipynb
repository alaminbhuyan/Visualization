```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load titanic datasets
titanic_df = sns.load_dataset("titanic")
titanic_df.head()
```

```python
titanic_df.shape
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df)
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex")
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,hue="sex",hue_order=["female","male"])
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,size="who")
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,size="who",sizes=(50,200))
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,size="who",sizes=(50,100),size_order=["child","woman","man"])
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,style="who")
plt.show()
```

```python
sns.scatterplot(x="age",y="fare",data=titanic_df,style="who",style_order=["child","woman","man"])
plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex",palette="hot")
plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = "who", markers= ['3','1',3])
plt.show()
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, alpha = 0.4)
plt.show()
```

```python
plt.figure(figsize=(16,9)) # figure size in 16:9 ratio

kwargs  =   {'edgecolor':"r",
             'facecolor':"k",
             'linewidth':2.7,
             'linestyle':'--',
            }

sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = "sex", sizes = (500, 1000), alpha = .7,  **kwargs)
```

```python
sns.set() #  set background 'darkgrid'
plt.figure(figsize = (16,9)) # figure size in 16:9 ratio

# create scatter plot
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex", palette = "magma",size = "who",
                sizes = (50, 300))
plt.title("Scatter Plot of Age and Fare", fontsize = 25) # title of scatter plot
plt.xlabel("Age", fontsize = 20) # x-axis label
plt.ylabel("Fare", fontsize = 20) # y-axis label
plt.savefig("Scatter Plot of Age and Fare") # save generated scatter plot at program location
plt.show() # show scatter plot
```

```python
sns.set() #  set background 'darkgrid'
plt.figure(figsize = (16,9)) # figure size in 16:9 ratio

# create scatter plot
sns.scatterplot(x = "who", y = "fare", data = titanic_df, hue = "alive", style = "alive", palette = "viridis",size = "who",sizes = (200, 500))

plt.title("Scatter Plot of Age and Fare", fontsize = 25) # title of scatter plot
plt.xlabel("Age", fontsize = 20) # x-axis label
plt.ylabel("Fare", fontsize = 20) # y-axis label
plt.savefig("Scatter Plot of Age and Fare") # save generated scatter plot at program location
plt.show() # show scatter plot
```
