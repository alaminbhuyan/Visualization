```python
global_warmin_df = pd.read_csv("D:\\JUPYTER2\\Seaborn\\Who_is_responsible_for_global_warming.csv")
global_warmin_df.head(3)
```

```python
# heatmap work only numerical data
global_warmin_df = global_warmin_df.drop(columns=["Country Code","Indicator Name","Indicator Code"],axis=1).set_index("Country Name")
global_warmin_df.head(2)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,cmap="coolwarm")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,cmap="coolwarm",center=10.0)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,cmap="coolwarm",vmin=0,vmax=21)
plt.show()
```

```python
## to remove the vmin,vmax in colorbar
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,cmap="copper",robust=True)
plt.show()
```

```python
# to add the value in heatmap
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True)
plt.show()
```

```python
# pass 2D Numpy array to annot parameter
annot_arr = np.arange(1,13).reshape(4,3) # create 2D numpy array with 4 rows and 3 columns
sns.heatmap(array_2d, annot= annot_arr)
plt.show()
```

```python
# fmp parameter - add text on heatmap cell
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype= str)

sns.heatmap(array_2d, annot = annot_arr, fmt="s") # s -string, d - decimal
plt.show()
```

```python
plt.figure(figsize=(15,10))
annot_kws={'fontsize':10,
           'fontstyle':'italic',
           'color':"k",
           'alpha':0.9,
           'rotation':"vertical",
           'verticalalignment':'center',
           'backgroundcolor':'w'}
sns.heatmap(data=global_warmin_df,annot=True,annot_kws=annot_kws)
plt.show()
```

```python
# linewidths parameter - divede each cell of heatmap
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True,linewidths=4)
plt.show()
```

```python
# linecolor parameter - change the color of heatmap line
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True,linewidths=4,linecolor="k")
plt.show()
```

```python
# hide color bar with cbar parameter
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True,linewidths=4,linecolor="k",cbar=False,cmap="winter_r")
plt.show()
```

```python
# orientation: To change the ‘horizontal’ or ‘vertical’ orientation of the color bar
# shrink: To change the size of the color bar
# extend: To change the end of the color bar like pointed or not. If you want pointed color bar both side  then passes value ‘both’, for left ‘min’ , right ‘max’ and ‘neither’ for no pointed color bar.
# extendfrac: To adjust the extension of the color bar. The ‘auto’ value adjust pointer automatically, ‘False’ value for no pointer and float value will help to adjust color bar pointer according to you.
# ticks: To change the ticks of the color bar, Pass list or numpy array of ticks
# drawedges: To draw lines (edges) on the color bar. Pass bool ‘True’ value for lines.

# change style and format of color bar with cbar_kws parameter

plt.figure(figsize=(14,14))

cbar_kws = {"orientation":"horizontal",
            "shrink":1,
            'extend':'both',
            'extendfrac':0.1,
            "ticks":np.arange(0,22),
            "drawedges":True,
           }

sns.heatmap(data=global_warmin_df, cbar_kws=cbar_kws)
plt.show()
```

```python
# change the shape of each cell in a square with squre parameter
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True,linewidths=3,linecolor="k",square=True)
plt.show()
```

```python
# change x-axis labels using xticklabels parameter
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True,linewidths=3,linecolor="k",square=True,xticklabels=np.arange(0,15))
plt.show()
```

```python
# change y-axis labels using yticklabels parameter

plt.figure(figsize=(15,10))

country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']

sns.heatmap(data=global_warmin_df, yticklabels = country_code)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,linewidths=3,linecolor="k",xticklabels=False,yticklabels=False,cbar=False)
plt.show()
```

```python
# multiple heatmaps using subplots

plt.figure(figsize=(30,10))

plt.subplot(1,3,1) # first heatmap
sns.heatmap(global_warmin_df,  cbar=False, linecolor="w", linewidths=2)

plt.subplot(1,3,2) # second heatmap
sns.heatmap(global_warmin_df,  cbar=False, linecolor="k", linewidths=2)

plt.subplot(1,3,3) # third heatmap
sns.heatmap(global_warmin_df,  cbar=False, linecolor="y", linewidths=2)

plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data=global_warmin_df,annot=True,linewidths=3,linecolor="k")
plt.title("Global warming",fontsize=15,color="w")
plt.xlabel("Years",fontsize=15,color="w")
plt.ylabel("Countrys Name",fontsize=15,color="w")
plt.show()
```

```python
# Heatmap with keyword arguments (kwargs) parameter

plt.figure(figsize = (15,10))
kwargs = {'alpha':.9,'linewidth':10, 'linestyle':'--', 'rasterized':False, 'edgecolor':'w',  "capstyle":'projecting',}

ax = sns.heatmap(data=global_warmin_df,**kwargs)
```

```python
# sns heatmap correlation
plt.figure(figsize=(16,9))

sns.heatmap(global_warmin_df.corr(), annot = True)
plt.show()
```

```python
# Upper triangle heatmap
plt.figure(figsize=(15,10))
# correlation matrix
corr_mx = global_warmin_df.corr()
# take lower correlation matrix
matrix = np.tril(corr_mx)
sns.heatmap(data=corr_mx,mask=matrix,annot=True,linecolor="k",linewidths=2)
plt.show()
```

```python
# Lower triangle heatmap
plt.figure(figsize=(15,10))
# correlation matrix
corr_mx = global_warmin_df.corr()
# take Upper correlation matrix
matrix = np.triu(corr_mx)
sns.heatmap(data=corr_mx,annot=True,mask=matrix)
plt.show()
```

```python
# set heatmap size
plt.figure(figsize= (15,10))

# create heatmap seaborn

cbar_kws = {"orientation":"vertical",
            "shrink":1,
            'extend':'min',
            'extendfrac':0.1,
            "ticks":np.arange(0,22),
            "drawedges":True,
           } # color bar keyword arguments

sns.heatmap(data=global_warmin_df, vmin = 0, vmax = 21, cmap="coolwarm", annot = True, linewidth = 2, cbar_kws=cbar_kws)

plt.title("Heatmap of 'Who is responsible for global warming'", fontsize = 25)
plt.xlabel("CO2 emissions (metric tons per capita) Per Year", fontsize = 20)
plt.ylabel("Country Name", fontsize = 20)
plt.show()
```

```python
# load dataset and create DataFrame ready to create heatmap
flights = sns.load_dataset("flights")
flights_df = flights.pivot("month", "year", "passengers")
# set heatmap size
plt.figure(figsize= (16,9))

# create heatmap seaborn

cbar_kws = {"shrink":.8,
           'extend':'both'}

sns.heatmap(flights_df, cmap="inferno", annot = True, fmt = 'd', linewidth = 2, cbar_kws=cbar_kws)

plt.title("Heatmap of 'Flights' Dataset", fontsize = 25)
plt.xlabel("Years", fontsize = 20)
plt.ylabel("Months", fontsize = 20)
plt.show()
```

```python
# set heatmap size
plt.figure(figsize= (16,9))

# create heatmap seaborn

cbar_kws = {"shrink":.8,
           'extend':'max',
           'extendfrac':.2,
           "drawedges":True}

sns.heatmap(flights_df.corr(), cmap="inferno", annot = True, linewidth = 2, cbar_kws=cbar_kws)

plt.title("Heatmap Correlation of 'Flights' Dataset", fontsize = 25)
plt.xlabel("Years", fontsize = 20)
plt.ylabel("Months", fontsize = 20)
plt.show()
```
