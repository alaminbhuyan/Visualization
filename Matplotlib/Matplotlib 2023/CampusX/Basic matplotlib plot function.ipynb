## Line Plot

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

more about line chart: https://github.com/alaminbhuyan/Visualization/blob/master/Matplotlib/Plot.ipynb

```python
# plotting a simple function

price = [48000,54000,57000,49000,47000,45000]

year = [2015,2016,2017,2018,2019,2020]

plt.plot(year, price)
plt.title(label="Year vs Price")
plt.xlabel(xlabel="Years")
plt.ylabel(ylabel="Price")
plt.show()
```

```python
df = pd.read_csv("dataset/sharma-kohli.csv")

df
```

```python
plt.figure(figsize=(8,7))
plt.plot(df['index'], df['RG Sharma'])
plt.title(label="Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel(xlabel="Season")
plt.ylabel(ylabel="Runs scored")
plt.show()
```

```python
# Plots multiple lines and add legend

plt.figure(figsize=(8,7))

plt.plot(df['index'], df['RG Sharma'], label="RG Sharma", marker="+")

plt.plot(df['index'], df['V Kohli'], label="V Kohli", marker="o")

plt.title(label="Rohit Sharma Vs Virat Kohli Career Comparison", 
          color="m", fontstyle="italic", fontsize=12, fontweight=5, pad=15)

plt.xlabel(xlabel="Season",color="white",fontsize=15,fontfamily="agency fb",backgroundcolor="black", loc='left')

plt.ylabel(ylabel="Runs scored",color="white",fontsize=15,fontfamily="agency fb",backgroundcolor="black", loc='top')

plt.legend()

plt.show()
```

```python
# Plots multiple lines and add legend

# "mo--" means color,marker,linestyle

plt.figure(figsize=(8,7))

plt.plot(df['index'], df['RG Sharma'], label="RG Sharma", color="g", marker="o", linestyle="--", linewidth = 3,
        markersize=10)

plt.plot(df['index'], df['V Kohli'], label="V Kohli", color="blue", marker="+", linestyle="--", linewidth = 3,
        markersize=10)

plt.title(label="Rohit Sharma Vs Virat Kohli Career Comparison", pad=15)

plt.xlabel(xlabel="Season")

plt.ylabel(ylabel="Runs scored")

plt.legend(loc=4)

# We set style so grid all ready created. If grid not present then use

# plt.grid()

plt.show()



# The strings
#     ``'upper left', 'upper right', 'lower left', 'lower right'``
#     place the legend at the corresponding corner of the axes/figure.

#     The strings
#     ``'upper center', 'lower center', 'center left', 'center right'``
#     place the legend at the center of the corresponding edge of the
#     axes/figure.
```
