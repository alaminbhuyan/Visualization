### 3D Surface Plots

```python
x = np.linspace(start=-10, stop=10, num=100)
y = np.linspace(start=-10, stop=10, num=100)
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
print(xx.shape)
print(yy.shape)
```

```python
# this is just a equation

z = xx**2 + yy**2

z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

# viridis

p = ax.plot_surface(X=xx, Y=yy, Z=z, cmap='rainbow')

fig.colorbar(p)

plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(X=xx, Y=yy, Z=z, cmap='viridis')

fig.colorbar(p)

plt.show()
```

```python
z = np.sin(xx) + np.log(xx)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(X=xx, Y=yy, Z=z, cmap='viridis')

fig.colorbar(p)

plt.show()
```

### Contour Plots

```python
x = np.linspace(start=-10, stop=10, num=100)
y = np.linspace(start=-10, stop=10, num=100)
xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx, yy, z, cmap='viridis')

fig.colorbar(p)

plt.show()
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx, yy, z, cmap='viridis')

fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p)
```
