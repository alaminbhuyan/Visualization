```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
# Dataset of 'Indian Artificial Intelligence Production (IAIP) Class"
 
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
 
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.figure(figsize=(10,7))
plt.bar(classes,class1_students)
plt.title("Different languages people",fontsize=15)
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of people",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.barh(classes,class1_students)
plt.title("Different languages people",fontsize=15)
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of people",fontsize=15)
plt.show()
```

# Use multiple parameters of plt.bar() method

```python
plt.figure(figsize=(10,7))
plt.bar(classes, class1_students, width = 0.3, align = "edge", color = "y",
       edgecolor = "m", linewidth = 3,
       label =" Class 1 Students") # visible=False this is hide the bar
plt.title("Different languages people",fontsize=15)
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of people",fontsize=15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "g",
       edgecolor = "r", linewidth = 5, alpha = 0.9, linestyle = "--",
       label =" Class 1 Students") # visible=False this is hide the bar
plt.title("Different languages people",fontsize=15)
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of people",fontsize=15)
plt.legend()
plt.show()
```
