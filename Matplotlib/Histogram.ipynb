```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45, (100))
py_students_age = np.random.randint(15,40, (100))
```

```python
print(ml_students_age)
print(py_students_age)
```

# Plotting histogram of machine learning students age.

```python
plt.figure(figsize=(10,8)) # figsize=(width,height)
plt.hist(ml_students_age)
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize=(10,8))
plt.hist(ml_students_age,bins=bins)
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize=(10,8))
plt.hist(ml_students_age,bins=bins,rwidth=0.8,color="m")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize=(10,8))
plt.hist(ml_students_age,bins=bins,rwidth=0.8,color="m",orientation="vertical",label="ML Students",
        histtype="bar")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize=(10,8))
plt.hist(ml_students_age,bins=bins,rwidth=0.8,color="m",orientation="vertical",label="ML Students",
        histtype="step")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis
plt.figure(figsize=(10,8))
plt.hist(ml_students_age,bins=bins,rwidth=0.8,color="y",orientation="horizontal",label="ML Students",
        histtype="bar")
plt.title("ML Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(10,8))
plt.hist([ml_students_age, py_students_age], bins=bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["m", "y"], label = ["ML Student", "Py Student"])
# plt.hist(ml_students_age,bins=bins,rwidth=0.8,color="y",orientation="horizontal",label="ML Students",
#         histtype="bar")
plt.title("ML &amp; Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
