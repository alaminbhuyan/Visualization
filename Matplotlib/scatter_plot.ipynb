```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randint(1,40,(80))
y = np.random.randint(1,17,(80))
z = np.random.randint(60,100,(80))
```

```python
print(x)
print(y)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y,color="g",marker="*",s=60,alpha=0.5,linewidths=5,edgecolors='c')
plt.title("Scatter plot")
plt.xlabel("alamin")
plt.ylabel("bhuyan")
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x,y,color="g",marker="*",s=60,alpha=0.5,linewidths=5,edgecolors='c')
plt.scatter(x,z,c="r",marker="*",s=60,alpha=0.5,linewidths=5,edgecolors="m")
plt.title("Scatter plot")
plt.xlabel("alamin")
plt.ylabel("bhuyan")
```
