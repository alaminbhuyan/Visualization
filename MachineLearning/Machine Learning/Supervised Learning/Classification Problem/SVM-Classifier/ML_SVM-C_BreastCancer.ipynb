```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```
d = load_breast_cancer()
df = pd.DataFrame(data=np.c_[d.data,d.target],columns=[list(d.feature_names)+["target"]])
df
```

```
df.shape
```

# Split data

```
features = df.iloc[:,0:-1]
level = df.iloc[:,-1]
```

```
features.shape
```

```
xtrain,xtest,ytrain,ytest = train_test_split(features,level,test_size=0.2,random_state=2020)
```

```
print('Shape of X_train = ', xtrain.shape)
print('Shape of y_train = ', ytrain.shape)
print('Shape of X_test = ', xtest.shape)
print('Shape of y_test = ', ytest.shape)
```

# Before features scaling
## Create Model

```
svm_rbf = SVC(kernel="rbf")
svm_rbf.fit(xtrain,ytrain)
svm_rbf.score(xtest,ytest)
```

# Apply features scaling

```
sc = StandardScaler()
```

```
x_train = sc.fit_transform(xtrain)
x_test = sc.transform(xtest)
```

```
svm_rbf_sc = SVC(kernel="rbf")
svm_rbf_sc.fit(x_train,ytrain)
svm_rbf_sc.score(x_test,ytest)
```

```
# Apply different kernel
svm_poly_sc = SVC(kernel="poly")
svm_poly_sc.fit(x_train,ytrain)
svm_poly_sc.score(x_test,ytest)
```

```
# Apply different kernel
svm_linear_sc = SVC(kernel="linear")
svm_linear_sc.fit(x_train,ytrain)
svm_linear_sc.score(x_test,ytest)
```

```
y_predict = svm_linear_sc.predict(X=xtest)
y_predict
```
