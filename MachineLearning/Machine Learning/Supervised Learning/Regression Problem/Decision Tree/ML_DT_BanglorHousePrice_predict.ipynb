```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```
df = pd.read_csv("/content/drive/My Drive/DataSets/bangalore house price prediction OHE-data.csv")
df.head()
```

```
df.shape
```

```
df['location_Marathahalli'].value_counts()
```

## Split the features and level

```
features = df.drop(labels=["price"],axis=1)
level = df["price"]
```

```
xtrain,xtest,ytrain,ytest = train_test_split(features,level,test_size = 0.3,random_state = 2021)
```

```
print("Shape of xtrain: ",xtrain.shape)
print("Shape of ytrain: ",ytrain.shape)
print("Shape of xtest: ",xtest.shape)
print("Shape of ytest: ",ytest.shape)
```

## Create the **Model**

```
regressior_model = DecisionTreeRegressor(criterion="mse") # mse = mean squaired error
regressior_model.fit(xtrain,ytrain)
```

```
print("Accuracy: ", round(regressior_model.score(X=xtest,y=ytest),ndigits=3)*100,"%")
```

```
# try another criterion="mae" its also called cost_function
regressior_model2 = DecisionTreeRegressor(criterion="mae") # mean absolute error
regressior_model2.fit(xtrain,ytrain)
print("Accuracy: ", round(regressior_model2.score(X=xtest,y=ytest),ndigits=3)*100,"%")
```

```
# try another criterion="friedman_mse"
regressior_model3 = DecisionTreeRegressor(criterion="friedman_mse",)
regressior_model3.fit(xtrain,ytrain)
print("Accuracy: ", round(regressior_model3.score(X=xtest,y=ytest),ndigits=3)*100,"%")
```

```
pred = regressior_model3.predict(xtest)
pred
```

## Apply k-fold **cross_validition**

```
# if we want apply cross-validition then we have to use feature & level
cvs = cross_val_score(estimator=regressior_model3,X=features,y=level,cv=10)
cvs
```

### Maximum & Minimum **Accuracy**

```
cvs.max(),cvs.min()
```

```
# Average Accuracy
cvs.mean()
```
