```
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```
df = pd.read_csv("/content/drive/My Drive/DataSets/bangalore house price prediction OHE-data.csv")
df.head()
```

```
df.shape
```

## Split the features and level

```
features = df.drop(labels=["price"],axis="columns")
level = df["price"]
```

```
print(features.shape)
print(level.shape)
```

```
xtrain,xtest,ytrain,ytest = train_test_split(features,level,test_size=0.2,random_state=2023)
```

```
print('Shape of X_train = ', xtrain.shape)
print('Shape of y_train = ', ytrain.shape)
print('Shape of X_test = ', xtest.shape)
print('Shape of y_test = ', ytest.shape)
```

### Create the model

```
rfr = RandomForestRegressor(n_estimators=200,criterion="mse")
```

```
rfr.fit(xtrain,ytrain)
rfr.score(X=xtest,y=ytest)
```

```
rfr2 = RandomForestRegressor(n_estimators=400,criterion="mse")
rfr2.fit(xtrain,ytrain)
rfr2.score(X=xtest,y=ytest)
```

## Apply k-fold cross_validtion

```
cvs = cross_val_score(estimator=rfr2,X=features,y=level,cv=10)
cvs
```

## To see max and min accuracy

```
cvs.max(),cvs.min()
```

```
# See average accuracy
cvs.mean()
```
