```
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
```

```
df = pd.read_csv("/content/drive/My Drive/DataSets/bangalore house price prediction OHE-data.csv")
df.head()
```

```
df.shape
```

```
df.isnull().sum().any()
```

# Split the features and level

```
features = df.drop(labels=["price"],axis="columns")
level = df["price"]
```

```
features.shape
```

```
xtrain,xtest,ytrain,ytest = train_test_split(features,level,test_size=0.2,random_state=215)
```

```
print('Shape of X_train = ', xtrain.shape)
print('Shape of y_train = ', ytrain.shape)
print('Shape of X_test = ', xtest.shape)
print('Shape of y_test = ', ytest.shape)
```

# Apply features scaling

```
sc = StandardScaler()
```

```
x_train = sc.fit_transform(X=xtrain)
x_test = sc.transform(X=xtest)
```

```
x_train
```

## Create the **model**

```
svm_rbf = SVR(kernel="rbf") # rbf = radial basis function
svm_rbf.fit(x_train,ytrain)
svm_rbf.score(x_test,ytest)
```

```
svm_linear = SVR(kernel="linear")
svm_linear.fit(x_train,ytrain)
svm_linear.score(x_test,ytest)
```

```
svm_poly = SVR(kernel="poly",degree=2)
svm_poly.fit(x_train,ytrain)
svm_poly.score(x_test,ytest)
```

```
pred_one = x_test[0]
```

```
svm_linear.predict([pred_one])
```

```
ytest.iloc[0]
```

```
y_pred = svm_linear.predict(x_test)
y_pred
```

```
from sklearn.metrics import mean_squared_error
```

```
mse = mean_squared_error(y_true=ytest,y_pred=y_pred)
rmse = np.sqrt(mse)
```
