## encoding

```python
encoder = LabelEncoder()
```

```python
x = dataset.apply(encoder.fit_transform)
x
```

```python
featurs = x.iloc[:,1:-1].values
featurs
```

```python
featurs1 = x.iloc[:,1:-1]
featurs1
```

```python
label = x.iloc[:,-1]
label
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(featurs1,label,test_size =.25,random_state=7)
```

```python
xtrain
```

```python
ytrain
```

## featurs scalling

```python
sc_x = StandardScaler()
```

```python
xtrain = sc_x.fit_transform(xtrain)
# xtest = sc_x.fit_transform(xtest)
xtest = sc_x.transform(xtest)
```

```python
xtrain
```

```python
xtest
```

# model selection

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
y_predict = model.predict(xtest)
```
