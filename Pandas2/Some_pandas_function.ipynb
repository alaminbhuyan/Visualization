# add()

```python
np.random.seed(25)
df = pd.DataFrame(np.random.rand(10,3),columns=['A','B','C'])
df
```

```python
df.iloc[-1] = np.nan
df
```

```python
df.add(other=10,fill_value=2)
```

```python
np.random.seed(10)
df2 = pd.DataFrame(np.random.rand(5,5),columns=['A','B','C','D','E'])
df2
```

```python
df3 = df.add(df2).copy()
```

```python
df3
```

```python
df3.add(other=2,axis='columns',fill_value=10)
```

# where()

```python
my_df = pd.read_csv('nba.csv')
```

```python
my_df.head(4)
```

```python
my_df.sort_values(by='Team',inplace=True)
```

```python
my_df.head(3)
```

```python
filters1 = (my_df['Team']=='Atlanta Hawks')
filters2 = my_df['Age']>24
my_df.where(filters1 & filters2)
```
