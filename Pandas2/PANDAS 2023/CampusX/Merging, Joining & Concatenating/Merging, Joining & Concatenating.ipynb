### merging with joining

```python
# We can use both of them

regs = pd.concat(objs=[nov, dec], ignore_index=True)

regs.head()
```

### inner join

```python
# one column should be same on the basis we apply merge function or join

# by default it apply inner join. if you want you can specify

students.merge(right=regs, on='student_id', how='inner')
```

```python
# Using pandas merge

pd.merge(left=students, right=regs, on='student_id', how='inner')
```

## Solve some problem statement

```python
# 1. find total revenue generated

total_revenue = regs.merge(right=courses, on='course_id', how='inner')['price'].sum()

total_revenue
```

```python
# 2. find month by month revenue

temp_df = pd.concat(objs=[nov, dec], keys=['Nov', 'Dec']).reset_index()

temp_df.head()
```

```python
temp_df.merge(right=courses, on='course_id').groupby(by='level_0')['price'].sum()
```

```python
# 3. Print the registration table
# cols -> student (name) -> course (name) -> course (price)

regs.merge(right=students, on="student_id").merge(right=courses, on="course_id")[['name','course_name','price']].head(10)
```

```python
# 4. Plot bar chart for revenue/course

regs.merge(right=courses, on='course_id').groupby(by='course_name')['price'].sum().plot(kind='bar')
```

```python
# 5. find students who enrolled in both the months

# here we have to find the common student id for nov, dec

(nov['student_id'], dec['student_id'])
```

```python
common_stu_id = np.intersect1d(ar1=nov['student_id'], ar2=dec['student_id'])

common_stu_id
```

```python
students[students['student_id'].isin(values=common_stu_id)]
```

```python
# 6. find course(name) that got no enrollment

# courses['course_id']
# regs['course_id']

course_id_list = np.setdiff1d(ar1=courses['course_id'], ar2=regs['course_id'])

course_id_list
```

```python
courses[courses['course_id'].isin(values=course_id_list)]
```

### `numpy.setdiff1d() function find the set difference of two arrays and return the unique values in arr1 that are not in arr2.`

```python
# 7. find students who did not enroll into any courses

student_id_list = np.setdiff1d(ar1=students['student_id'], ar2=regs['student_id'])

students[students['student_id'].isin(values=student_id_list)]
```
