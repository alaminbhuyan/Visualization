```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(88)
```

```python
np.random.randn(10, 4)
```

```python
df = pd.DataFrame(data={'A': np.linspace(start=1, stop=10, num=10)})

df2 = pd.DataFrame(data=np.random.randn(10, 4), columns=list("BCDF"))
```

```python
df
```

```python
df2
```

```python
df3 = pd.concat(objs=[df, df2], axis=1)

df3
```

```python
df3.iloc[3,3] = np.nan
df3.iloc[0,2] = np.nan

df3
```

### For highlighting Maximum values

```python
df3.style.highlight_max()
```

```python
# change color

df3.style.highlight_max(color="pink")
```

### For highlighting minimum values

```python
df3.style.highlight_min()
```

### Highlight Null values

```python
df3.style.highlight_null(color="red")
```

`set_na_rep(): Along with highlighting the missing values, they may be represented as “nan”. You can change the representation of these missing values using the set_na_rep() function. This function can also be chained with any styler function but chaining it with highlight_null will provide more details.`

```python
df3.style.format(na_rep="OutofScope").highlight_null(color="orange")
```

### Create Heatmap within dataframe

There are few parameters you can pass to this function to further customize the output generated:

cmap: By default, the “PuBu” colormap is selected by pandas You can create a custom matplotlib colormap and pass it to the camp parameter.

axis: Generating heat plot via rows or columns criteria, by default: columns
        
text_color_threshold: Controls text visibility across varying background colors.

```python
df3.style.background_gradient()
```

### Table Properties

```python
df3.style.set_properties(**{'border': '1.3px solid green',
                          'color': 'magenta'})
```

### Create Bar charts

Just as the heatmap, the bar charts can also be plotted within the dataframe itself. The bars are plotted in each cell depending upon the axis selected. By default, the axis=0 and the plot color are also fixed by pandas but it is configurable. To plot these bars, you simply need to chain the “.bar()” function to the styler object.

```python
df3.style.bar()
```

### Control precision

```python
df3.style.format(precision=3)
```

### Add Captions

```python
df3.style.set_caption("This is Analytics Vidhya Blog").format(precision=3).background_gradient()
```

### Hiding Index or Column

```python
df3.style.hide(axis="index")
```

### Control display values

```python
df3.style.format("{:.3%}")
```

```python
df3.style.format("{:.3%}", na_rep="&&")
```
